```python
import pandas as pd

df_public = pd.read_csv("cleaned_data.csv")
df_public.head(5)
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from collections import Counter
import re

plt.rc('font', family='NanumGothic') 
%matplotlib inline

#     (:  )
font_path = 'C:/tobigs/etc/Fonts/NanumGothic.ttf'

# 
stopwords = [
    "", "", "", "","", "^^", "^^;", "", "", "", "", "", "", 
    "", "", "", "", "", "", "", "", "", "...", "", "", "", "", 
    "", "", "", "", "", "", "", "", "", "", 
    "", "", "", "", "", "", "", " ", " ", 
    "", "", "", "", "", "", "", "", "", "", "", 
    "", "", "", "", "", "", "", "", "", "", "", "", "", 
    "", "", "", "", "", "", "", "", "", "", "", "", 
    "", "", "", "", "", "", "", "", "", "", "", 
    "", "", "", "", "", "", "", "", "", "", "", 
    "", "", "", "", "", "", "", "", "", "", "", 
    "", "", "", "", "", "", "", "", "", "", "", 
    "", "", "", "", " ", "", " ", "", "", "", 
    "", "", "", "", "", "", "", "  ", "", "", 
    "", " ", "", "", " ", "", "", "", "", "", 
    "", "", " ", "", "", "", "", "", "", "", "", 
    "", "", "", "", "", "", "", "", "", "", "", 
    "", "", "", "", "", "", "", "", "", "", 
    "", "  ", "  ", " ", " ", " ", " ", " ", 
    " ", "", "", "", "", "", " ", " ", " ", 
    " ", " ", " ", "", "", " ", " ", "  ", 
    " ", " ", " ", " ", "", "", "", "",
    "", "", "", "", "", "", "", "", "",
    "", "", "", "", "", "", "", "", "", "", "",
    "", "", "", "", "", "", "", "", "", "", "",
    "", "", "", "", "", "", "", "", "", "", "", "", "",
    "newsblock", "", "tv", "", "", "Newsblock", "", "", "", "", "", ""
]

#   
def remove_stopwords(text):

    text = re.sub(r'\d+', '', text)  #  
    words = re.findall(r'\b\w+\b', text)
    filtered_words = [word for word in words if word not in stopwords]
    return " ".join(filtered_words)

#  location content   
locations = df_public['location'].unique()

for location in locations:
    location_content = df_public[df_public['location'] == location]['content'].dropna()
    combined_text = " ".join(location_content.apply(remove_stopwords))
    
    #   
    word_counts = Counter(combined_text.split())
    
    #   ( )
    wordcloud = WordCloud(
        font_path=font_path,  #    
        width=800, 
        height=400, 
        background_color='white'
    ).generate_from_frequencies(word_counts)
    
    #  
    plt.figure(figsize=(10, 5))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.title(f"{location}   ")
    plt.axis("off")
    plt.show()

```
