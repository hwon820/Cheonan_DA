```python
import urllib.request
import json


# KNU ê°ì • ì‚¬ì „ ë‹¤ìš´ë¡œë“œ
url = "https://raw.githubusercontent.com/park1200656/KnuSentiLex/master/KnuSentiLex/data/SentiWord_info.json"
urllib.request.urlretrieve(url, "SentiWord_info.json")

# KNU ê°ì • ì‚¬ì „ ë¡œë“œ ë° íŒŒì‹±
def load_knu_senti_lexicon():
    sentiment_dict = {}
    with open("SentiWord_info.json", "r", encoding="utf-8") as f:
        data = json.load(f)
        for entry in data:
            word = entry['word']
            polarity = int(entry['polarity'])
            sentiment_dict[word] = polarity
    return sentiment_dict

# ê°ì • ì‚¬ì „ ë¡œë“œ
knu_sentiment_dict = load_knu_senti_lexicon()

# í…ìŠ¤íŠ¸ ì „ì²˜ë¦¬ ë° í† í°í™”
def preprocess_text(text):
    from konlpy.tag import Okt
    okt = Okt()
    tokens = okt.morphs(text)
    return tokens

# ê°ì • ë¶„ì„ ìˆ˜í–‰
def sentiment_analysis(text, sentiment_dict):
    tokens = preprocess_text(text)
    sentiment_scores = {"Positive": 0, "Negative": 0}
    
    word_scores = {}
    
    for word in tokens:
        score = sentiment_dict.get(word, 0)
        word_scores[word] = score
        
        if score > 0:
            sentiment_scores["Positive"] += score
        elif score < 0:
            sentiment_scores["Negative"] += score
    
    total_score = sentiment_scores["Positive"] + sentiment_scores["Negative"]
    
    if total_score > 0:
        overall_sentiment = "Positive"
    elif total_score < 0:
        overall_sentiment = "Negative"
    else:
        overall_sentiment = "Neutral"
    sentiment_scores['total_sentiment'] = overall_sentiment
    
    return sentiment_scores, word_scores
```
