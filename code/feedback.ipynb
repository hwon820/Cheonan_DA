```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# 사용자 정의 한국어 불용어 리스트
stopwords = set(['블로그에서는 다음과 같은','평가하였습니다','있어','강아지','애견','언급하였습니다','있다', '있어서', '있었', '했다', '것', '수', '하는', '으로', '합니다', '했습니', '이', '있습니다'])

# 한글 폰트 경로 설정
font_path = '/usr/share/fonts/truetype/nanum/NanumGothicBold.ttf'  # 나눔고딕 Bold 폰트 경로

# 데이터 준비
text = ' '.join(private_facil['pos'])

# 워드 클라우드 객체 생성
wc = WordCloud(font_path=font_path, width=800, height=400, background_color='white', stopwords=stopwords).generate(text)

# 워드 클라우드 표시
plt.figure(figsize=(10, 5))
plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```
