### 유의미한 피드백 필터링

#### 시립 시설

```python
import json
import pandas as pd
import numpy as np
import os
```

```python
with open('Data/merged_location_improvements.json') as f:
    data = json.load(f)
```

```python
facil_issues = []

# JSON 데이터 파싱
for keys, entries in data.items():
    for entry in entries:
        facil_issues.append({
            "title": keys,
            "url": entry["url"],
            "raw_review": entry["review"],
            "issues": entry["issues"],
        })


df = pd.DataFrame(facil_issues)
```

```python
df.head()
```

```python
len(df)
```

```python
len(df[df['issues']=="개선점 없음"])
```

```python
# isses가 "개선점 없음"인 데이터 제거
filtered_df = df[(df['issues'] != "개선점 없음") & (df['issues'] != "Rate limit exceeded, please try later.")]
filtered_df
```

```python
len(filtered_df)
```

### 직접 확인 후 제거
25개의 데이터만 남아 개선점 확인 후 개선점이 없는 데이터를 제거하였다.

제거할 인덱스 번호
> 8, 10, 14, 18, 20, 21, 22

```python
# 확인용
filtered_df.iloc[22].issues
```

```python
# 해당 인덱스 번호의 행을 삭제
rows_to_drop = [8, 10, 14, 18, 20, 21, 22]
filtered_df = filtered_df.drop(filtered_df.index[rows_to_drop])

filtered_df
```

```python
filtered_df.reset_index(drop=True, inplace=True)
```

```python
filtered_df.head()
```

#### 사설 시설

```python
with open('Data/private_facil_positives.json') as f:
    data2 = json.load(f)
```

```python
len(df2)
```
