```python
import json
import pandas as pd
```

### 데이터 살펴보기

```python
with open('Data/blog.json', 'r') as f:
    json_data = json.load(f)

print(type(json_data))
```

```python
print(list(json_data.keys()))
```

```python
json_data['도솔광장 반려인 쉼터'][0].keys()
```

```python
posts = []

# JSON 데이터 파싱
for location, entries in json_data.items():
    for entry in entries:
        posts.append({
            "location": location,
            "title": entry["title"],
            "url": entry["url"],
            "postdate": entry["postdate"],
            "content": entry["text"]
        })


df = pd.DataFrame(posts)

# 날짜 형식 변환
df['postdate'] = pd.to_datetime(df['postdate'], format='%Y%m%d')
```

```python
df.shape
```

### 중복 데이터 제거

```python
# df_no_duplicates = df.drop_duplicates()

# 열 기준 중복 행 제거 'Content' 기준
df_no_duplicates_specific_column = df.drop_duplicates(subset=['content'])
print(df_no_duplicates_specific_column)
```

```python
df = df_no_duplicates_specific_column
```

### 관련없는 글 제거

```python
# 제거할 키워드 리스트
keywords = [
    "광고", "협찬", "공지", "이벤트", "일보", " 기자",
    "발표", "제품 제공", "제품제공", "소정의", "꿀팁이당대표"
]
```

```python
# 제거할 데이터 확인용
pattern = '|'.join(keywords)
df[df['content'].str.contains(pattern, case=False, na=False)]
```

```python
# 제거할 데이터 확인용
df[df['url'].str.contains("news", case=False, na=False)]
```

```python
# 필터링
pattern = '|'.join(keywords)
filtered_df = df[~df['content'].str.contains(pattern, case=False, na=False)]
filtered_df = filtered_df[~filtered_df['url'].str.contains('news', case=False, na=False)]

print(filtered_df)
```
