```python
import json
import pandas as pd
```

```python
with open('Data/private_facil_blog.json', 'r') as f:
    json_data = json.load(f)

print(type(json_data))
```

```python
print(list(json_data.keys()))
```

```python
posts = []

# JSON 데이터 파싱
for location, entries in json_data.items():
    for entry in entries:
        posts.append({
            "location": location,
            "title": entry["title"],
            "url": entry["url"],
            "postdate": entry["postdate"],
            "content": entry["text"]
        })


df = pd.DataFrame(posts)

# 날짜 형식 변환
df['postdate'] = pd.to_datetime(df['postdate'], format='%Y%m%d')
```

```python
df.shape
```

```python
# 열 기준 중복 행 제거 'Content' 기준
df_no_duplicates_specific_column = df.drop_duplicates(subset=['content'])
print(df_no_duplicates_specific_column)
```
